pseudo code :  

<pre>
Topological Sort with STACK:    
    Create STACK  
    Create visited <- false  
    for each vertex v:  
        if (!visited[v])  
            DFS(v,visited,STACK)  
    while (!STACK.empty)  
        STACK.pop  
  
DFS(v,visited,STACK):  
    visited[v] <- true  
    for i in Adj[v]  
        if (!visited[i])  
            DFS  
    while (!STACK.empty)  
        STACK.push(v)  

가능한 위상정렬 순서 개수 : 42개
풀이과정 : 진입 간선이 없는 a부터 시작한다. a의 진출 간선을 제거하면 b,d로 나뉜다. b로 가면 그 다음은 c로 이어진다. d로 가면 같은 과정을 거쳐서 d -> e -> f -> g -> h 이거나 d -> e -> f -> h -> g 이다. 8개의 정점 중 a는 첫번째 자리로 고정이고 나머지 7개에서 d ~ h를 나열하는 두 가지 경우가 있다. 그리고 7개의 자리 중 정해진 5개의 자리를 제외한 b,c가 들어갈 두 자리를 고르는 경우의 수가 7C2=21이다. 따라서 2 x 21 = 42로 총 42가지이다.
